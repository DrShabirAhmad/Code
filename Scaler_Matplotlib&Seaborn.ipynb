```python
top3_pub =data["Publisher"].value_counts().index[:3]
top3_genre = data["Genre"].value_counts().index[:3]
top3_platform = data["Platform"].value_counts().index[:3]
top3_pub, top3_genre, top3_platform
```

```python
top3_data =data.loc[(data["Publisher"].isin(top3_pub)) & (data["Platform"].isin(top3_platform)) & (data["Genre"].isin(top3_genre))]
top3_data.head()
```

```python
sns.countplot(x="Publisher", data= top3_data)
```

```python
sns.countplot(x="Publisher", data= top3_data, width=.4)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x="Publisher", hue="Genre", data=top3_data, width=.5)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x="Publisher", hue = "Platform", data=top3_data, width=.4)
plt.title("Publisher wise Platform usage")
plt.show()
```

```python
fig = plt.figure(figsize=(15,10))
sns.scatterplot(x='NA_Sales', y='EU_Sales', data=top3_data)
fig.suptitle('Sales Relationship')
plt.show()
```

```python
fig = plt.figure(figsize=(15,10))

plt.subplot(2,3,1)
sns.scatterplot(x='NA_Sales', y='EU_Sales', data=top3_data)

plt.subplot(2,3,3)
sns.scatterplot(x='NA_Sales', y='JP_Sales', data=top3_data, color="r")

fig.suptitle('NU sales comparison')
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,10))
axes[0,0].scatter(top3_data['NA_Sales'], top3_data['EU_Sales'])
axes[0,1].scatter(top3_data['NA_Sales'], top3_data['JP_Sales'])
axes[1,0].scatter(top3_data['NA_Sales'], top3_data['Other_Sales'])
axes[1,1].scatter(top3_data['NA_Sales'], top3_data['Global_Sales'])
plt.show()
```

```python
plt.figure(figsize=(20,12)).suptitle("NA Sales vs regions",fontsize=20)

plt.subplot(2, 2, 1)
sns.scatterplot(x='NA_Sales', y='EU_Sales', data=top3_data)
plt.title('NA vs EU Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('EU', fontsize=12)

plt.subplot(2, 2, 2)
sns.scatterplot(x='NA_Sales', y='JP_Sales', data=top3_data, color='red')
plt.title('NA vs JP Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('JP', fontsize=12)

plt.subplot(2, 2, 3)
sns.scatterplot(x='NA_Sales', y='Other_Sales', data=top3_data, color='green')
plt.title('NA vs Other Region Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('Other', fontsize=12)

plt.subplot(2, 2, 4)
sns.scatterplot(x='NA_Sales', y='Global_Sales', data=top3_data, color='orange')
plt.title('NA vs Global Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('Global', fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(20,12)).suptitle("NA Sales vs regions",fontsize=20)


# Using a 2x3 subplot
plt.subplot(2, 3, 1)
sns.scatterplot(x='NA_Sales', y='EU_Sales', data=top3_data)
plt.title('NA vs EU Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('EU', fontsize=12)

plt.subplot(2, 3, 3)
sns.scatterplot(x='NA_Sales', y='JP_Sales', data=top3_data, color='red')
plt.title('NA vs JP Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('JP', fontsize=12)


plt.subplot(2, 3, 4)
sns.scatterplot(x='NA_Sales', y='Other_Sales', data=top3_data, color='green')
plt.title('NA vs Other Region Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('Other', fontsize=12)

plt.subplot(2, 3, 6)
sns.scatterplot(x='NA_Sales', y='Global_Sales', data=top3_data, color='orange')
plt.title('NA vs Global Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('Global', fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(20,12)).suptitle("NA Sales vs regions",fontsize=20)


# Using a 2x3 subplot
plt.subplot(2, 3, 1) # rows, columns, position
sns.scatterplot(x='NA_Sales', y='EU_Sales', data=top3_data)
plt.title('NA vs EU Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('EU', fontsize=12)

plt.subplot(2, 3, 3)
sns.scatterplot(x='NA_Sales', y='JP_Sales', data=top3_data, color='red')
plt.title('NA vs JP Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('JP', fontsize=12)

# Countplot of publishers
plt.subplot(1,3,2)
sns.countplot(x='Publisher', data=top3_data)
plt.title('Count of games by each Publisher', fontsize=12)
plt.xlabel('Publisher', fontsize=12)
plt.ylabel('Count of games', fontsize=12)

plt.subplot(2, 3, 4)
sns.scatterplot(x='NA_Sales', y='Other_Sales', data=top3_data, color='green')
plt.title('NA vs Other Region Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('Other', fontsize=12)

plt.subplot(2, 3, 6)
sns.scatterplot(x='NA_Sales', y='Global_Sales', data=top3_data, color='orange')
plt.title('NA vs Global Sales', fontsize=12)
plt.xlabel('NA', fontsize=12)
plt.ylabel('Global', fontsize=12)

plt.show()
```
