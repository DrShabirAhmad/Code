```python
bcr.bar_chart_race(df,
                   figsize=(5,3), dpi=100,
                   label_bars=False,
                   period_label={'x':.99, 'y': .1, 'ha':'right', 'color':'red'}, title="Country wise Covid-19 Deaths")
```

```python
bcr.bar_chart_race(df, period_fmt='%b %-d, %Y', title="Country wise Covid-19 Deaths")
```
