# Marginal & Conditional Probability

```python
pd.crosstab(df.Product, df.Gender, values=df.Gender, aggfunc='count', normalize='all',  margins=True, margins_name='Total')
# What is the probability of a male customer buying a KP781 treadmill?   Ans = 0.1833
```

```python
pd.crosstab(df.Product, df.Gender, values=df.Gender, aggfunc='count', normalize=True)
```

```python
pd.crosstab(index=df.Product, columns= df.Gender)
```

```python
pd.crosstab(df.Product, df.Gender, normalize='all')
```

```python
pd.crosstab(df.Product, df.Gender, normalize='index')
```

```python
pd.crosstab(df.Product, df.Gender, normalize='columns', margins=True, margins_name='Total')
```

```python
pd.crosstab(df.Product, df.Gender, margins=True)
```

```python
pd.crosstab(df.Product, df.Gender, margins=True, margins_name='Total', normalize='all')
```

```python
pd.crosstab(df.Product, df.Gender, margins=True, margins_name='Total')
```

```python
pd.crosstab([df.Product, df.Gender], df.MaritalStatus, margins=True, margins_name='Total')
```
